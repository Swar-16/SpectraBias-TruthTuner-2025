```python
import html
```

```python
def decode_html_entities(text):
    """
    Convert HTML entities and special characters in the input text to plain text.
    Handles common HTML entities like &amp;#039;, &amp;lsquo;, &amp;rsquo;, &amp;quot;, &amp;amp;, etc.
    """
    if not isinstance(text, str):
        return text
    return html.unescape(html.unescape(text))


df['content'] = df['content'].apply(decode_html_entities)
```

# **3rd Editing**

*   Manually removed "iframe", "script", "div", "style"  tags
*   Manually removed parts of text that were promotional (like, share, subscribe, publisher name etc.)
*   Manually removed "views are personal"






```python
#Removed JS (function(){...})(...); blocks
def remove_js_functions(text):
    if not isinstance(text, str):
        return text
    pattern = re.compile(r'\(function\s*\([^\)]*\)\s*\{.*?\}\s*\)\s*\([^\)]*\)\s*;?', re.DOTALL)
    return pattern.sub('', text)

df['content'] = df['content'].apply(remove_js_functions)
```

```python
#Removed https and www. links
def remove_hyperlinks(text):
    if not isinstance(text, str):
        return text
    # Matches http, httpsand www. links
    pattern = r'https?://\S+|www\.\S+'
    return re.sub(pattern, '', text)

df['content'] = df['content'].apply(remove_hyperlinks)
```

```python
#Removed twitter links
def remove_pic_twitter_links(text):
    if not isinstance(text, str):
        return text
    return re.sub(r'pic\.twitter\.com/\S+', '', text)

df['content'] = df['content'].apply(remove_pic_twitter_links)
```

```python
#Removed Advertisement text
#then manually removed Advertisements
def remove_advertisement(text):
    if not isinstance(text, str):
        return text
    return re.sub(r'\bADVERTISEMENT\b', '', text, flags=re.IGNORECASE).strip()

df['content'] = df['content'].apply(remove_advertisement)
```

```python
df.to_csv('df_edited_1.csv', index=False)
```
