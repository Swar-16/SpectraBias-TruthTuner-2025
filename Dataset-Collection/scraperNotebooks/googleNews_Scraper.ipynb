```python
#searching function
def get_articles(search) :
  gn = GoogleNews(lang='en', country='IND')
  search = gn.search(search)
  articles = search['entries']
  news = []
  for i in articles :
    #['title', 'title_detail', 'links', 'link', 'id', 'guidislink', 'published', 'summary', 'summary_detail', 'source', 'sub_articles']
    article = {"title" : i["title"], "published" : i["published"], "summary" : i["summary"], "source" : i["source"]["title"]}
    news.append(article)
  return news


allData = get_articles("Ind vs Pak")
```
